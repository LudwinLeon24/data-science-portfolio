```python
# Modified datetime columns
df['FECHA LLEGADA'] = pd.to_datetime(df['FECHA LLEGADA'], errors='coerce', dayfirst=True)
df['FECHA PEDIDO'] = pd.to_datetime(df['FECHA PEDIDO'], errors='coerce', dayfirst=True)
df['DIA'] = pd.to_datetime(df['DIA'], errors='coerce', dayfirst=True)

# Modifies columns to numeric types

cols_a_convertir = [
    'PRECIO', 'BOLSA', 'STICKER', 'STICKER PRECIO', 'TRANSPORTE',
    'COSTO DE UNIDAD', 'PRECIO', 'PRECIO A VENTA', 'PRECIO FINAL',
    'REINVERSION', 'LLAMPI', 'UTILIDADES', 'REINVERSION.V',
    'LLAMPI.V', 'UTILIDADES', 'PAGO AL BANCO', 'TOTAL'
]

for col in cols_a_convertir:   #generamos un bucle para convertir las columnaas seleccionadas a tipo numerico
    df[col] = (
        df[col]
        .astype(str)     # nos aseguramos de que todo sea texto
        .str.replace('$', '', regex=False)  # eliminamos el simbolo dolar y regex evita que pandas lo interprete como una expresion regular
        .str.replace(',', '', regex=False)  # eliminamos las comas
        .str.strip() # eliminamos los espacion al incio y al final
        .replace('', np.nan)  # eliminamos las secciones vacioas por Nan
        .astype(float) # convertimos a tipo float
    )

display(df.info())
display(df.describe().round(2))
```
